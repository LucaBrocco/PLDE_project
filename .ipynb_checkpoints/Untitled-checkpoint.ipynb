```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Resident population.csv")
data 
```

```python
ages = data['Age'][:100]
nums = data['Total'][:100]

ages_label = ages.copy()
for i in range(len(ages_label)):
    if i%10 != 0:
        ages_label[i] = ''
plt.bar(ages,nums,width=0.5,tick_label = ages_label)
plt.xlabel('Age')
plt.ylabel('Population')
plt.grid(alpha = 0.4)
plt.savefig('ita_pop_hist.png')
plt.show()
```

```python
ages = data['Age'][:100]
nums = data['Total'][:100]

ages_label = ages.copy()
for i in range(len(ages_label)):
    if i%10 != 0:
        ages_label[i] = ''
plt.bar(ages[:19],nums[:19],width=0.5,tick_label = ages_label[:19],label ='Youth',color = 'lightblue')
#plt.grid(alpha = 0.4)
plt.bar(ages[19:65],nums[19:65],width=0.5,tick_label = ages_label[19:65],label ='Adults',color = 'orange')
#plt.grid(alpha = 0.4)
plt.bar(ages[65:],nums[65:],width=0.5,tick_label = ages_label[65:],label ='Elderly',color = 'green')
plt.xlabel('Age')
plt.ylabel('Population')

tt = np.arange(0,len(ages),1)
plt.bar(tt,tt,tick_label=ages_label)
plt.grid(alpha = 0.4)
plt.legend()
plt.savefig('ita_pop_hist_byclass.png')

plt.show()
```
