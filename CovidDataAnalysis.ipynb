```python
import os 
import numpy as np
import pandas as pd
```

```python
folder_path = '/home/luca3/Desktop/PoD/plde/project/covid_data/COVID-19/dati-province/'

dataframe = pd.DataFrame()
for date in pd.date_range(start = '2020-02-24', end = '2020-12-31'):
    data = date.strftime('%Y%m%d')
    file_path = folder_path + 'dpc-covid19-ita-province-' + data + '.csv'
    df = pd.read_csv(file_path)
    df = df[df['sigla_provincia'] == 'PD']
    dataframe = pd.concat([dataframe,df])
```

```python
cols = ['data','totale_casi']
data = dataframe[cols]

idx = np.arange(0,np.shape(data)[0],1)
data.index = idx
data.to_csv('PD_covid_data.csv')
```
